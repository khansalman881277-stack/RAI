# 📊 Fairness-Aware Machine Learning for Diabetes Prediction

## 🚀 Project Overview

This project demonstrates how to build a machine learning model for diabetes prediction while addressing bias and fairness issues using the AI Fairness 360 (AIF360) toolkit.

Machine learning models can unintentionally learn biases from data, leading to unfair predictions for different demographic groups. This project focuses on building a fair and responsible ML pipeline.

---

## 📂 Dataset

* Pima Indians Diabetes Dataset
* Features include:

  * Pregnancies, Glucose, Blood Pressure, BMI, Age, etc.
* A synthetic **Gender** feature is created to simulate fairness analysis.

---

## ⚙️ Technologies Used

* Python
* Pandas & NumPy
* Scikit-learn
* Matplotlib
* imbalanced-learn (SMOTE)
* AI Fairness 360 (AIF360)

---

## 🔍 Project Pipeline

### 1. Data Preprocessing

* Load dataset
* Create protected attribute (Gender)
* Perform train-test split

### 2. Handling Imbalanced Data

* Apply SMOTE to balance class distribution

### 3. Feature Scaling

* Standardize numerical features (excluding Gender)

### 4. Model Training

* Train Logistic Regression model
* Achieved accuracy: **~74.5%**

### 5. Fairness Evaluation

Measured using:

* Statistical Parity Difference (SPD)
* Disparate Impact (DI)

---

## ⚖️ Bias Mitigation

Applied **Reweighing algorithm** from AIF360 to reduce bias in the dataset.

---

## 📊 Results

| Metric | Before Mitigation | After Mitigation |
| ------ | ----------------- | ---------------- |
| SPD    | -0.0307           | ~0.0000          |
| DI     | 0.94              | 1.00             |

✅ Bias significantly reduced after mitigation
✅ Fairness improved across demographic groups

## 💡 Key Insights

* Machine learning models can contain hidden bias
* Fairness metrics are critical in sensitive domains like healthcare
* Bias mitigation improves fairness without major accuracy loss
